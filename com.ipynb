```python
dataset.dropna(inplace=True)
```

```python
df = dataset[["track_id","track_name","artists","track_genre","popularity"]].copy()
```

```python
# Remove duplicate tracks by name + artist
df = df.drop_duplicates(subset=['track_name', 'artists'])

# Optional: aggregate genres / popularity if multiple entries exist
df = df.groupby(['track_name','artists']).agg({
    'track_genre': lambda x: ', '.join(set(x)),
    'popularity': 'max'  # or 'mean'
}).reset_index()
```

```python
df.shape
df.head(10)
```

```python
df['name_df'] = df['track_name'].apply(lambda x : x.split())
```

```python
df['artists_df'] = df['artists'].apply(lambda x : x.split())
```

```python
df['genre_df'] = df['track_genre'].apply(lambda x : x.split())
```

```python
df
```

```python

def simple_clean(i):
    df[i] = df[i].apply(
        lambda x: [re.sub(r'[^A-Za-z0-9]', '',''.join(x)).lower().strip()]
    )
    
```

```python
simple_clean('artists_df')
simple_clean('name_df')
simple_clean('genre_df')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
df['combined_features'] = df['name_df']+df['artists_df'] + df['genre_df']
```

```python
df["combined_features"] = df['combined_features'].apply(lambda x: ' '.join(x))
```

```python
vector = TfidfVectorizer(max_features=5000,stop_words="english")
matrix = vector.fit_transform(df["combined_features"])
```

```python
vector.get_feature_names_out()
```

```python
df['track_name_clean'] = df['track_name'].str.lower().str.strip()
```

```python
from difflib import get_close_matches
```

```python
def recommend(song_name, top_n=10):
    song_name_clean = song_name.lower().strip()
    
    matches = df[df['track_name_clean'] == song_name_clean]
    if matches.empty:
        close = get_close_matches(song_name_clean, df['track_name_clean'].tolist(), n=1, cutoff=0.6)
        if not close:
            return f"Song '{song_name}' not found in the database."
        matches = df[df['track_name_clean'] == close[0]]
    
    index = matches.index[0]
    song_vector = matrix[index]
    sim_scores = cosine_similarity(song_vector, matrix).flatten()
    
    # get top indices (excluding itself)
    top_indices = np.argsort(sim_scores)[-top_n-1:-1][::-1]
    
    # ✅ return a dataframe instead of printing
    return df.iloc[top_indices][["track_name", "artists", "track_genre", "popularity"]]
```

```python
recommend("Comedy")
```
